```python
import pandas as pd
```

```python
teams = pd.read_csv("resources/nfl_teams.csv")
```

```python
team_csv = 'resources/team_stats/' + 'arizona-cardinals' + '.csv'

df = pd.read_csv(team_csv)

new_columns = df.title.to_list()
data = [df.home.to_list()]

just_home = pd.DataFrame(data,columns= new_columns)
just_home['team'] = 'Arizona Cardinals'
just_home
```

```python
#setting an empty list
frames = []

for team in teams.nfl_teams:
    team_name = team.lower().replace(' ', '-')
    team_csv = 'resources/team_stats/' + team_name + '.csv'

    df = pd.read_csv(team_csv)

    new_columns = df.title.to_list()
    data = [df.home.to_list()]

    just_home = pd.DataFrame(data,columns= new_columns)
    just_home['team'] = team
    frames.append(just_home)
```

```python
all_teams_df = pd.concat(frames)
all_teams_df = all_teams_df.reset_index(drop=True)

all_teams_df
```

```python
all_teams_df.to_csv('resources/all_teams_stats.csv',index=False)
```
