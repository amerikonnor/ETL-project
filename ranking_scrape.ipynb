```python
# Dependencies
import os
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
url = 'https://www.profootballnetwork.com/nfl-power-rankings-end-of-regular-season-week-17-2020/'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
rankings_section = soup.find('div', class_='zox-post-body left zoxrel zox100')
```

```python
rankings_section.prettify()
```

```python
rankings_html = rankings_section.find_all('h4')
```

```python
rankings_html
```

```python
for ranking in rankings_html:
    print(ranking.text.split())

```

```python
rankings_html[6]
```

```python
rank = []
team = []
record = []

for ranking in rankings_html:
    current_team = ranking.text.split()
    if len(current_team) == 4:
        current_rank = current_team[0]
        current_rank = int(current_rank[:-1])
        rank.append(current_rank)
        current_team_name = current_team[1] + ' ' + current_team[2]
        team.append(current_team_name)
        current_record = current_team[3]
        record.append(current_record)
    elif len(current_team) == 5:
        current_rank = current_team[0]
        current_rank = int(current_rank[:-1])
        rank.append(current_rank)
        current_team_name = current_team[1] + ' ' + current_team[2] + ' ' + current_team[3]
        team.append(current_team_name)
        current_record = current_team[4]
        record.append(current_record)
    elif len(current_team) == 3:
        current_rank = current_team[0]
        current_rank = int(current_rank[:-1])
        rank.append(current_rank)
        current_team_name = 'Philadelphia Eagles'
        team.append(current_team_name)
        current_record = '(4-10-1)'
        record.append(current_record)
    
        
```

```python
rank_df = pd.DataFrame({'rank':rank,'team':team,'record':record})
rank_df.sort_values(by = 'team', inplace = True)
rank_df.set_index('rank', inplace = True)
rank_df
```

```python
rank_df.to_csv('rankings.csv')
```
