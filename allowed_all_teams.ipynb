```python
import pandas as pd
```

```python
teams = pd.read_csv("resources/nfl_teams.csv")
```

```python
team_csv = 'resources/team_stats/' + 'arizona-cardinals' + '.csv'

df = pd.read_csv(team_csv)

new_columns = df.title.to_list()

for column in new_columns:
    new_columns[new_columns.index(column)] = 'allowed ' + column



data = [df.opponent.to_list()]

just_allowed = pd.DataFrame(data,columns= new_columns)
just_allowed['team'] = 'Arizona Cardinals'
just_allowed
```

```python
#setting an empty list
frames = []

for team in teams.nfl_teams:
    team_name = team.lower().replace(' ', '-')
    team_csv = 'resources/team_stats/' + team_name + '.csv'

    df = pd.read_csv(team_csv)

    new_columns = df.title.to_list()

    for column in new_columns:
        new_columns[new_columns.index(column)] = 'allowed ' + column

    data = [df.opponent.to_list()]
    
    just_allowed = pd.DataFrame(data,columns= new_columns)
    just_allowed['team'] = team
    frames.append(just_allowed)
```

```python
allowed_all_teams_df = pd.concat(frames)
allowed_all_teams_df = allowed_all_teams_df.reset_index(drop=True)

allowed_all_teams_df
```

```python
allowed_all_teams_df.to_csv('resources/allowed_all_teams_stats.csv',index=False)
```
