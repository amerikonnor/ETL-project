```python
import pandas as pd
```

```python
teams = pd.read_csv("resources/nfl_teams.csv")
```

```python
team_csv = 'resources/team_stats/' + 'arizona-cardinals' + '.csv'

df = pd.read_csv(team_csv)

new_columns = df.title.to_list()

for column in new_columns:
    new_columns[new_columns.index(column)] = 'allowed ' + column



data = [df.opponent.to_list()]

just_allowed = pd.DataFrame(data,columns= new_columns)
just_allowed['team'] = 'Arizona Cardinals'
just_allowed
```

```python
#setting an empty list
frames = []

new_columns = [
    'total_first_downs',
    'first_downs_rush_pass_penalty',
    'third_down_conversions',
    'fourth_down_conversions',
    'total_offensive',
    'offense_plays_avg',
    'total_rushing',
    'rushing_plays_avg',
    'total_passing',
    'passing_comp_att_int_avg',
    'sacks',
    'field_goals',
    'touchdowns',
    'touchdowns_rush_pass_returns_def',
    'avg_possesion',
    'turnover_ratio'
]

#because i copied and pasted but need allowed
for column in new_columns:
        new_columns[new_columns.index(column)] = 'allowed_' + column

for team in teams.team_name:
    team_name = team.lower().replace(' ', '-')
    team_csv = 'resources/team_stats/' + team_name + '.csv'

    df = pd.read_csv(team_csv)

    data = [df.opponent.to_list()]
    
    just_allowed = pd.DataFrame(data,columns= new_columns)
    just_allowed['team_id'] = teams.nfl_teams.to_list().index(team) + 1
    frames.append(just_allowed)
```

```python
allowed_all_teams_df = pd.concat(frames)
allowed_all_teams_df = allowed_all_teams_df.reset_index(drop=True)

allowed_all_teams_df
```

```python
cols = allowed_all_teams_df.columns.to_list()
cols = cols[-1:] + cols[:-1]
allowed_all_teams_df = allowed_all_teams_df[cols]
allowed_all_teams_df
```

```python
allowed_all_teams_df.to_csv('resources/allowed_all_teams_stats.csv',index=False)
```
